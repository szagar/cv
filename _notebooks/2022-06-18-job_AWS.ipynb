### Glassdoor
* Traverse a tree (BSF or DSF)
* Given 2 sorted arrays, merge them. Follow up, ensure only 1 of each element is saved in the merged array aka no duplicates.
    * You can use the merge sort algorithm to answer this question
    * Answer=set(Array1+Array2)
* Given reverse polish notation, return the integer value of the expression aka "5,3,+,2,/" would be 5 + 3 = 8 and then 8 / 2 = 4.
    * Use a stack. Push values onto the stack. When the token is an operator, pop the 2 values from the stack and apply the operator, pushing the result back onto the stack. Make sure to put the values on the correct sides of the operator. At the end, if it's a valid postfix notation or "reverse polish notation", there will be a single value on the stack. That's the answer.
* Add spaces to a sentence given a sentence and dictionary. "iloveamazon" should return "i love amazon" for example
* check if string has balanced brackets, firstly with "()" only and thereafter adding all other brackets such as "[]{}<>".
    * [valid parenthesis](https://leetcode.com/problems/valid-parentheses/)
